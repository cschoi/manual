## Homebrew를 이용해 Python을 설치한 분들을 위한 Neuron 패키지 설치법  (Yosemite 사용자)

### - 설치에 엄청난 도움을 주신 이성호님께 큰 감사 드립니다.
- 박성진님. 자세하게 설치법 올려주셨는데 제가 못따라가서 죄송합니다 ㅠㅠ

### 1. 필요한 패키지 설치

> brew install mercurial

> brew install autoconf

> brew install libtool

> brew install automake


### 2. 설치 디렉토리 만들기

> sudo mkdir /Applications/NEURON-7.4

> sudo mkdir /Applications/NEURON-7.4/iv

> sudo mkdir /Applications/NEURON-7.4/nrn

> mkdir nrn64

> cd nrn64

### 3. nrn64 폴더 내에 neuron 패키지 source code 다운로드

> hg clone http://www.neuron.yale.edu/hg/neuron/iv

> hg clone http://www.neuron.yale.edu/hg/neuron/nrn

### 4. iv 폴더 내 패키지 빌드

> cd iv

> sh build.sh

##### 마지막에 나오는 warning 메시지는 error가 아니니 무시해도 됨.

### 5. 본인 시스템 체크

> ./config.guess

##### 명령어 결과는 다음처럼 출력 (사용하는 OS에 따라 다를 수도 있음)

> x86_64-apple-darwin14.1.0

##### 여기서 나온 x86_64-apple-darwin14.1.0 은 iv와 nrn 패키지 make 시 build 와 host 옵션에서 사용함.

### 6. iv 폴더 내 패키지 컴파일

> ./configure --prefix=/Applications/NEURON-7.4/iv --build=x86_64-apple-darwin14.1.0 --host=x86_64-apple-darwin14.1.0

> make
> sudo make install

### 7. nrn 폴더 내 패키지 빌드

> cd ../nrn

> sh build.sh

### 8. nrn 폴더 내 패키지 컴파일

> ./configure --prefix=/Applications/NEURON-7.4/nrn --with-iv=/Applications/NEURON-7.4/iv --with-x --x-includes=/usr/X11/include/ --x-libraries=/usr/X11/lib/ --with-nrnpython=/usr/local/Cellar/python/2.7.9/Frameworks/Python.framework/Versions/2.7/Resources/Python.app/Contents/MacOS/Python --host=x86_64-apple-darwin14.1.0 --build=x86_64-apple-darwin14.1.0

##### 주의 - brew로 python을 설치한 사람은 꼭 nrnpython= 다음에 저 경로를 써줘야 함.

#### 8.1. cfg 파일 만들기 (이유는 잘 모름. 이성호님께 물어볼 예정)

> vi ~/.pydistutils.cfg

[install]

prefix=

##### :wq! 를 써서 저장 후 종료

> make

> sudo make install

> sudo make install after_install

#### 8.2. 컴파일 후 .cfg 파일 지우기

> rm ~/.pydistutils.cfg

##### 여기까지 하면 Application에 NEURON 패키지 관련 아이콘들이 생성되는 것을 볼 수 있음.

### 9. python 모듈로 연동

> cd ~/nrn64/nrn/src/nrnpython

> echo $PYTHONPATH

> echo "export PYTHONPATH=$PYTHONPATH:/usr/local/lib/python2.7/site-packages" >> ~/.bash_profile

배쉬 종료 후 다시 켬 (PATH 적용)

#### 9.1. virtualenv 활성화 

본인의 경우 brainpy라는 버츄얼환경을 만들어 NEURON을 그 안에서만 사용하고 싶은 욕망이 있었음. 그냥 표준 설치 된 파이썬에서 사용하고 싶으신 분은 source 명령어를 빼고 바로 setup.py로 넘어가야함.
> source ~/Study/brainpy/bin/activate

> python setup.py install
